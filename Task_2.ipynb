```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train_path = "/content/train.csv"
test_path = "/content/test.csv"
if not os.path.exists(train_path) or not os.path.exists(test_path):
    print("Error: One or more dataset files are missing. Please upload train.csv and test.csv.")
else:

    train_df = pd.read_csv(train_path)
    test_df = pd.read_csv(test_path)

    print(train_df.info())
    print(train_df.isnull().sum())
    print(train_df.head())

    train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
    train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)
    train_df.drop(columns=['Cabin'], inplace=True)

    print(train_df.isnull().sum())

    sns.set_style("whitegrid")

    plt.figure(figsize=(6, 4))
    sns.countplot(x="Survived", data=train_df, palette="pastel")
    plt.title("Survival Count (0 = No, 1 = Yes)")
    plt.xlabel("Survived")
    plt.ylabel("Count")
    plt.show()

    plt.figure(figsize=(6, 4))
    sns.countplot(x="Pclass", hue="Survived", data=train_df, palette="Set2")
    plt.title("Survival by Passenger Class")
    plt.xlabel("Passenger Class")
    plt.ylabel("Count")
    plt.legend(title="Survived", labels=["No", "Yes"])
    plt.show()

    plt.figure(figsize=(6, 4))
    sns.countplot(x="Sex", hue="Survived", data=train_df, palette="husl")
    plt.title("Survival by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.legend(title="Survived", labels=["No", "Yes"])
    plt.show()

```
